```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler
```

```python
# Load data
df = pd.read_csv("../data/creditcard.csv")
```

```python
# 1. Class Imbalance Analysis
print("Class Distribution:")
print(df['Class'].value_counts())
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```

```python
# 2. Distributions of 'Time' and 'Amount'
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df['Time'], ax=ax[0], kde=True, bins=50)
ax[0].set_title("Distribution of Transaction Time")
sns.histplot(df['Amount'], ax=ax[1], kde=True, bins=50)
ax[1].set_title("Distribution of Transaction Amount")
plt.show()
```

```python
# 3. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of Features")
plt.show()

# 4. Outlier Detection (Boxplots)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale('log')  # Log scale for better visualization
plt.title("Transaction Amount by Fraud Class")
plt.show()
```

```python
# 5. Feature Scaling (Preview)
scaler = RobustScaler()
df['Scaled_Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['Scaled_Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
print("\nScaled Amount/Time Sample:")
print(df[['Amount', 'Scaled_Amount', 'Time', 'Scaled_Time']].head())
```
